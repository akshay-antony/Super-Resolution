```python
%load_ext autoreload
%autoreload 2
import numpy as np
import torch
from range_to_pcd import RangeToPCD
import os
```

```python
device = torch.device("cpu")
```

```python
range_2_pcd = RangeToPCD(image_rows=64,
                        ang_start_y=16.6,
                        fov_y=33.2,
                        max_range=80,
                        min_range=2,
                        device=device)
high_res_image = np.load("./bag/carla_ouster_range_image.npy")
if not os.path.exists("./bag/pcd_ouster/"):
    os.makedirs("./bag/pcd_ouster/")
```

```python
count = 0
for i in range(high_res_image.shape[0]):
    curr_image = torch.from_numpy(high_res_image[i].squeeze(-1))
    curr_image = np.expand_dims(curr_image, 0)
    curr_image = np.expand_dims(curr_image, 1)
    curr_image /= 80
    curr_pcd = range_2_pcd.convert_range_to_pcd(curr_image).squeeze(0)
    curr_pcd = curr_pcd.cpu().numpy()
    np.save("./bag/pcd_ouster/" + str(i), curr_pcd)
    print(i)
```
